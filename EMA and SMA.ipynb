```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Define URL
url = "https://finance.yahoo.com/quote/%5EGSPC/history"

# Headers to mimic a browser visit
headers = {"User-Agent": "Mozilla/5.0"}

# Get webpage content
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Find table with historical data
table = soup.find("table")

# Convert table into DataFrame
df = pd.read_html(str(table))[0]

# Display first few rows
print(df.head())
```

```python
import pandas as pd

# Rename columns to proper names
df.columns = ["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]

# Convert 'Date' column to datetime format
df["Date"] = pd.to_datetime(df["Date"])

# Convert numerical columns to float (ignore errors for now)
for col in ["Open", "High", "Low", "Close", "Adj Close"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Convert 'Volume' to integer (if needed)
df["Volume"] = pd.to_numeric(df["Volume"], errors="coerce").astype("Int64")

# Sort Data by Date in ascending order
df = df.sort_values("Date").reset_index(drop=True)

# Display cleaned data
print(df.head())
df.to_csv("S&P500_10.csv", index=False)
```

**SMA and EMA**

```python
import pandas as pd
import yfinance as yf
ticker = "SBUX"
start_date = "2015-02-28"
end_date = "2025-03-01"    # Target date

# Download Starbucks stock df for the specified range
df = yf.download(ticker, start=start_date, end=end_date)

df = df.loc[:, [('Close', 'SBUX'), ('High', 'SBUX'), ('Low', 'SBUX'), ('Open', 'SBUX'), ('Volume', 'SBUX')]].copy()

# Flatten MultiIndex column names
df.columns = ['Close', 'High', 'Low', 'Open', 'Volume']

# Reset index to make 'Date' a column if it's the index
df.reset_index(inplace=True)
```

```python
# df
df["SMA20"] = df["Close"].rolling(window=20).mean()
df["SMA50"] = df["Close"].rolling(window=50).mean()

# Calculate Exponential Moving Averages (EMA)
df["EMA20"] = df["Close"].ewm(span=20, adjust=False).mean()
df["EMA50"] = df["Close"].ewm(span=50, adjust=False).mean()

# Display the last few rows
print(df.tail())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))

plt.plot(df['Date'], df['Close'], label='Close Price')
plt.plot(df['Date'], df['SMA20'], label='SMA20')
plt.plot(df['Date'], df['SMA50'], label='SMA50')
plt.legend()
plt.title('SBUX Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate Buy/Sell Signals
df['Buy_Signal'] = (df['SMA20'] > df['SMA50']) & (df['SMA20'].shift(1) <= df['SMA50'].shift(1))
df['Sell_Signal'] = (df['SMA20'] < df['SMA50']) & (df['SMA20'].shift(1) >= df['SMA50'].shift(1))

# Plotting
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Close'], label='Close Price', color='gray', linewidth=1)
plt.plot(df['Date'], df['SMA20'], label='SMA20', color='blue', linewidth=1.2)
plt.plot(df['Date'], df['SMA50'], label='SMA50', color='orange', linewidth=1.2)

# Plot Buy signals
plt.scatter(df.loc[df['Buy_Signal'], 'Date'],
            df.loc[df['Buy_Signal'], 'Close'],
            label='Buy Signal',
            marker='^',
            color='green',
            s=80)

# Plot Sell signals
plt.scatter(df.loc[df['Sell_Signal'], 'Date'],
            df.loc[df['Sell_Signal'], 'Close'],
            label='Sell Signal',
            marker='v',
            color='red',
            s=80)

# Chart formatting
plt.title('SBUX Stock Price with SMA Buy/Sell Signals', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Make sure Date is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Calculate EMAs
df['EMA20'] = df['Close'].ewm(span=20, adjust=False).mean()
df['EMA50'] = df['Close'].ewm(span=50, adjust=False).mean()

# Generate Buy/Sell Signals
df['Signal'] = 0
df.loc[df['EMA20'] > df['EMA50'], 'Signal'] = 1
df.loc[df['EMA20'] < df['EMA50'], 'Signal'] = -1

df['Position'] = df['Signal'].diff()

# Plotting
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Close'], label='Close Price', color='black', alpha=0.6)
plt.plot(df['Date'], df['EMA20'], label='EMA20', color='blue', linewidth=1.5)
plt.plot(df['Date'], df['EMA50'], label='EMA50', color='orange', linewidth=1.5)

# Plot Buy Signals
plt.scatter(df[df['Position'] == 2]['Date'],
            df[df['Position'] == 2]['Close'],
            label='Buy Signal',
            marker='^',
            color='green',
            s=100)

# Plot Sell Signals
plt.scatter(df[df['Position'] == -2]['Date'],
            df[df['Position'] == -2]['Close'],
            label='Sell Signal',
            marker='v',
            color='red',
            s=100)

plt.title('Starbucks (SBUX) - EMA Crossover Strategy')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
